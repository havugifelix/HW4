### Mask creation & Template matching
The function point extract points from left(reference image) by defining extracting a K by k template matrix from  the orginal image at different location in the image. Thus template will be used to find the matching point from the second image using convolution. 

```python

def point_extract(img,startx,starty,k):
    mask=[]
    for i in range(startx,starty+k):
        for t in range(starty,starty+k):
            mask.append(img[i,t])
    mask=np.reshape(mask,(k,k))
    point=img[startx+int((k+1)/2),int(starty+(k+1)/2)]
    return mask,point
```
